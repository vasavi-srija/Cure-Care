```python
import pandas as pd
column_list = ["Breathing Problem","Fever","Dry Cough","Sore throat","Running Nose","Headache","Contact with COVID Patient","COVID-19"]
dataset = pd.read_csv('./Covid_Dataset.csv',usecols=column_list)
dataset.head()
```

```python
import numpy as np
def split_data(dataset):
    new_data_array = np.random.permutation(len(dataset))
    test_data_index = new_data_array[:int(len(dataset)*0.2)]
    train_data_index = new_data_array[int(len(dataset)*0.2):]
    return dataset.iloc[train_data_index], dataset.iloc[test_data_index]
```

```python
train_data, test_data = split_data(dataset)
```

```python
X_train = train_data[['Breathing Problem','Fever', 'Dry Cough', 'Sore throat', 'Running Nose','Headache','Contact with COVID Patient']].to_numpy()
X_test = test_data[['Breathing Problem','Fever', 'Dry Cough', 'Sore throat', 'Running Nose','Headache','Contact with COVID Patient']].to_numpy()
Y_train = train_data[['COVID-19']].to_numpy().ravel()
Y_test = test_data[['COVID-19']].to_numpy().ravel()
```

```python
from sklearn.linear_model import LogisticRegression
import pickle
clf = LogisticRegression()
clf.fit(X_train, Y_train)
pickle.dump(clf, open("model.pkl", 'wb'))
```

```python
inputFeatures = [1, 0, 0, 0, 1, 0, 1]
loaded_model = pickle.load(open("model.pkl", 'rb'))
result =loaded_model.predict([inputFeatures])[0]
if(result):
    print("Positive")
else:
    print("Negative")
```

```python
acc = loaded_model.score(X_test, Y_test) 
print('Accuracy of the model: ', acc*100, ' %')
```
