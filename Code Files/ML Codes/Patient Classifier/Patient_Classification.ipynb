```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv('COVID-19.csv')
data
```

```python
data.info()
```

```python
data_df = pd.DataFrame(data,columns=['age','gender','body temperature','weakness','breathing problem','drowsiness','pain in chest','stroke or reduced immunity','diabetes','high blood pressue','Loss of sense of smell','Corona result'])
data_df
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
def data_split(data, ratio):
    np.random.seed(42)
    shuffled = np.random.permutation(len(data))
    test_set_size = int(len(data) * ratio)
    test_indices = shuffled[:test_set_size]
    train_indices = shuffled[test_set_size:]
    return data.iloc[train_indices], data.iloc[test_indices]
```

```python
train, test = data_split(data_df, 0.2)
```

```python
train
```

```python
test
```

```python
X_train = train[['age','gender','body temperature','weakness','breathing problem','drowsiness','pain in chest','stroke or reduced immunity','diabetes','high blood pressue','Loss of sense of smell']].to_numpy()
X_test = test[['age','gender','body temperature','weakness','breathing problem','drowsiness','pain in chest','stroke or reduced immunity','diabetes','high blood pressue','Loss of sense of smell']].to_numpy()

Y_train = train[['Corona result']].to_numpy().ravel()
Y_test = test[['Corona result']].to_numpy().ravel()
```

```python
model = LogisticRegression()
```

```python
import pickle
model.fit(X_train,Y_train)
pickle.dump(model, open("model2.pkl", 'wb'))
```

```python
inputFeatures = [45,1,100,1,1,1,1,1,1,1,1]
loaded_model = pickle.load(open("model2.pkl", 'rb'))
infProb =loaded_model.predict([inputFeatures])
if infProb == 0:
  print("No Severity")
elif infProb == 1:
  print("Mild Severity")
elif infProb == 2:
  print("High Severity")
```

```python
accuracy = model.score(X_test,Y_test)
accuracy 
```
